```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_test = pd.read_csv('test_set.csv', sep=';', names=['Feature', 'Result'])
df_train = pd.read_csv('train_set.csv', sep=';', names=['Feature', 'Result'])
```

```python
# (1 point) Display the scatter plot for the training set data.
plt.scatter(df_train.iloc[:, 0], df_train.iloc[:, 1], c='blue')
```

```python
#Data transformation
X_train = df_train.iloc[:, 0].to_numpy().reshape(-1,1)
y_train = df_train.iloc[:, 1].to_numpy().reshape(-1,1)
X_test = df_test.iloc[:, 0].to_numpy().reshape(-1,1)
y_test = df_test.iloc[:, 1].to_numpy().reshape(-1,1)
```

```python
def sort_two_lists(list1, list2):
    # Merging data by columns
    merged = np.concatenate((list1, list2), axis=1)
    # Sorting by arguments that are in the 0th column of the array 'merged'.
    sorted_list = merged[merged[:, 0].argsort()]
    return(sorted_list)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def poly_reg(X, Y, deg, dataset_length):
    #Creation of features for a given degree
    regression = PolynomialFeatures(degree=deg)
    features = regression.fit_transform(X)
    #Training a regression model
    polynomial_regression = LinearRegression()
    polynomial_regression.fit(features, Y)
    
    #print(features[139,:])#print(features[138,:])#print(features[137,:])#print(regression[40, :])    
    
    return regression, polynomial_regression

y_prediction_res = []
mse_test_all = []
models = []
parameter = []
parameters = []
features_out = []

def poly_train(i, X_train, X_test, y_train, y_test, dataset_length):
    #Wczytanie modelu 
    features, model = poly_reg(X_train, y_train, i, dataset_length)
    
    features_out.append(features)
    
    #linspace potrzebny do stworzenia wykresu funkcji
    #space = np.linspace(0, 10, X_train.size).reshape(-1, 1)
    #preds = model.predict(features.transform(space))
    
    #Predykcja dla modelu o stopniu "i"
    preds_train = model.predict(features.transform(X_train))
    preds_test = model.predict(features.transform(X_test))
    y_prediction_res.append(preds_test)
    
    #Wyliczeniu błedów średniokwadratowych
    mse_train = mean_squared_error(y_train, preds_train)
    mse_test = mean_squared_error(y_test, preds_test)
    
    #Tworzenie listy błedów do zad3
    mse_test_all.append(mse_test)

    #różnica = predykcja dla zbioru testowego – rzeczywista wartość zbioru testowego
    error_value = np.subtract(preds_test, y_test)
    

    #histogram gęstości rozkładu błędów między wartościami przewidywanymi
    #przez model dla zbioru testowego, a faktycznymi wartościami zbioru testowego
    plt.hist(error_value, density=True) 
    plt.ylabel('Density')
    plt.xlabel('Error value');
    plt.title(f'Histogram of errors for {i} degree')
    plt.tight_layout()
    plt.show()
    
    
    models.append(model)
    parameter.append(model.intercept_)
    parameters.append(model.coef_)
    
    print('Degree: ', i)
    print('Train error: {}\nTest error: {}'.format(mse_train, mse_test))
    
```

Wytrenować modele regresji na danych ze zbioru treningowego dla różnych stopni wielomianów (co najmniej od 1 do 12 stopnia włącznie). Dla każdego modelu naszkicować histogram gęstości rozkładu błędów między wartościami przewidywanymi przez model dla zbioru testowego, a faktycznymi wartościami zbioru testowego, tj.

```python
for i in  range(1, 13):
    poly_train(i, X_train, X_test, y_train, y_test, len(X_train) + len(X_test))
```

```python
print("The lowest error is for degree equal to: ", mse_test_all.index(min(mse_test_all)) + 1)
```
