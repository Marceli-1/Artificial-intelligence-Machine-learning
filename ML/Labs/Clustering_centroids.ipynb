```python
def k_means_clustering(data, n_clusters, init, epochs):
    clustering = KMeans(n_clusters=n_clusters, random_state=1, init=init, max_iter=epochs).fit(data)
    assignment = clustering.labels_
    centroids = clustering.cluster_centers_
    
    fig, ax = plt.subplots(figsize=(10, 7))
    points = ax.scatter(data[:, 0], data[:, 1], c=assignment)
    centres = ax.scatter(centroids[:, 0], centroids[:, 1], c='red')
    legend = ax.legend(*points.legend_elements(), title='Cluster No.')
    ax.add_artist(legend)
    plt.show()
    return clustering, assignment, centroids
```

```python
make_cluster, assigment, centroids = k_means_clustering(data, 4, 'k-means++', 300)
```
